###<font color = "#5ccda7">**1. ÍNDICES - casos geolocalizados**

```python
db.casos_epidemiologicos.create_index(
    [("DT_NOTIFIC", ASCENDING)],
    name="idx_casos_data_notificacao"
)

db.casos_epidemiologicos.create_index(
    [("ID_UNIDADE", ASCENDING)],
    name="idx_casos_id_unidade"
)

db.casos_epidemiologicos.create_index(
    [("NO_FANTASIA", ASCENDING)],
    name="idx_casos_nome_hospital"
)

db.casos_epidemiologicos.create_index(
    [("CS_SEXO", ASCENDING)],
    name="idx_casos_sexo"
)

db.casos_epidemiologicos.create_index(
    [("NU_IDADE_N", ASCENDING)],
    name="idx_casos_idade"
)

db.casos_epidemiologicos.create_index(
    [("HOSPITALIZ", ASCENDING)],
    name="idx_casos_hospitalizacao"
)

db.casos_epidemiologicos.create_index(
    [("ID_MUNICIPIO", ASCENDING)],
    name="idx_casos_municipio"
)

db.casos_epidemiologicos.create_index(
    [("DT_NOTIFIC", ASCENDING), ("ID_UNIDADE", ASCENDING)],
    name="idx_casos_data_unidade"
)

db.casos_epidemiologicos.create_index(
    [("LATITUDE", ASCENDING), ("LONGITUDE", ASCENDING)],
    name="idx_casos_lat_long"
)
```

###<font color = "#5ccda7"> **2. ÍNDICES - dados climáticos**

```python
db.dados_climaticos.create_index(
    [("DT_MEDICAO", ASCENDING)],
    name="idx_clima_data_medicao"
)

db.dados_climaticos.create_index(
    [("PRECIP_TOTAL", DESCENDING)],
    name="idx_clima_precipitacao"
)

db.dados_climaticos.create_index(
    [("TEMP_MED", DESCENDING)],
    name="idx_clima_temperatura"
)

db.dados_climaticos.create_index(
    [("DT_MEDICAO", ASCENDING), ("PRECIP_TOTAL", DESCENDING)],
    name="idx_clima_data_precipitacao"
)

db.dados_climaticos.create_index(
    [("DT_MEDICAO", ASCENDING), ("TEMP_MED", DESCENDING)],
    name="idx_clima_data_temperatura"
)
```

### <font color = "#5ccda7">**3. ÍNDICES - risco de inundação**

```python
db.risco_inundacao.create_index(
    [("CLASSE", ASCENDING)],
    name="idx_risco_classe"
)

db.risco_inundacao.create_index(
    [("geometry", ASCENDING)],
    name="idx_risco_geometry"
)
```

### <font color = "#5ccda7">**4. ÍNDICES - vunerabilidade habitacional**

```python
db.vulnerabilidade_habitacional.create_index(
    [("NOME_AREA", ASCENDING)],
    name="idx_vulnerabilidade_nome_area"
)
```
